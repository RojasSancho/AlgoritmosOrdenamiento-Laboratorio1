### Ejercicio 2
Probar que f(n) = 5n² + 100n + 20 es Θ(n²).

Cota superior (O):\
Si se toma n ≥ 1, se cumple que:\
f(n) = 5n² + 100n + 20 ≤ 125n²\
Por lo tanto, f(n) pertenece a O(n²).

Cota inferior (Ω):\
Tomando n ≥ 1, se cumple que:\
f(n) = 5n² + 100n + 20 ≥ 5n²\
Por lo tanto, f(n) pertenece a Ω(n²).

Los términos 100n y 20 crecen más lento que n², por eso no afectan el comportamiento asintótico de la función cuando n es grande.

Conclusión:\
Como f(n) pertenece a O(n²) y f(n) pertenece a Ω(n²), entonces se puede decir que f(n) también pertenece a Θ(n²).
